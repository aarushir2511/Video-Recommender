# **1) CONTENT BASED FILTERING**

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

merged_df = pd.merge(unified_interactions, final_all_posts_df, left_on='post_id', right_on='id', how='inner')

liked_videos = merged_df[~merged_df['liked_at'].isnull()]

def combine_features(row):
    # Combine text-based features into a single string
    features = [
        row.get('keywords', ''),
        row.get('topics_of_video', ''),
        row.get('emotions', ''),
        row.get('psycological_view_of_video', ''),
        row.get('visual_elements_of_video', '')
    ]
    return ' '.join(map(str, features))

final_all_posts_df['combined_features'] = final_all_posts_df.apply(combine_features, axis=1)


tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(final_all_posts_df['combined_features'])


similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)
def recommend_videos_for_user(user_id, num_recommendations=10):

    user_liked_posts = liked_videos[liked_videos['user_id'] == user_id]['post_id'].tolist()

    if not user_liked_posts:
        return f"User {user_id} has not liked any videos."

    # Create a set to store recommendations
    recommended_videos = set()

    for liked_post in user_liked_posts:
        # Get the index of the liked video in the similarity matrix
        liked_post_index = final_all_posts_df[final_all_posts_df['id'] == liked_post].index[0]
        similar_indices = similarity_matrix[liked_post_index].argsort()[-num_recommendations-1:-1][::-1]

        # Add the most similar videos to recommendations
        for idx in similar_indices:
            recommended_video_id = final_all_posts_df.iloc[idx]['id']
            # Exclude videos the user has already liked
            if recommended_video_id not in user_liked_posts:
                recommended_videos.add(recommended_video_id)

        if len(recommended_videos) >= num_recommendations:
            break

    recommended_videos_df = final_all_posts_df[final_all_posts_df['id'].isin(recommended_videos)].head(num_recommendations)
    return recommended_videos_df[['id', 'title', 'keywords', 'topics_of_video']]

user_id_input = 1
recommended_videos = recommend_videos_for_user(user_id_input)
print(recommended_videos)
```

# **2) Collaborative filtering**

```python
def create_user_item_matrix(interactions_df, posts_df):
    # Initialize an empty dataframe with user ids as rows and post ids as columns
    user_item_matrix = pd.DataFrame(index=interactions_df['user_id'].unique(), columns=posts_df['id'].unique(), data=0.0)

    # Iterate over interactions and update the matrix
    for _, row in interactions_df.iterrows():
        user_id = row['user_id']
        post_id = row['post_id']


        if pd.notna(row['viewed_at']):
            user_item_matrix.at[user_id, post_id] = 1.0
        if pd.notna(row['liked_at']):
            user_item_matrix.at[user_id, post_id] = 2.0
        if pd.notna(row['inspired_at']):
            user_item_matrix.at[user_id, post_id] = 3.0
        if pd.notna(row['rating_percent']):
            user_item_matrix.at[user_id, post_id] = float(row['rating_percent'])

    return user_item_matrix


def compute_user_similarity(user_item_matrix):
    user_item_matrix_filled = user_item_matrix.fillna(0)
    similarity_matrix = cosine_similarity(user_item_matrix_filled)
    user_similarity_df = pd.DataFrame(similarity_matrix, index=user_item_matrix.index, columns=user_item_matrix.index)

    return user_similarity_df


def recommend_posts(user_id, user_similarity_df, user_item_matrix, num_recommendations=5):
    # Get similar users based on cosine similarity
    similar_users = user_similarity_df[user_id].sort_values(ascending=False).index[1:]

    post_recommendations = {}

    # Loop through similar users
    for similar_user in similar_users:
        similar_user_posts = user_item_matrix.loc[similar_user]

        # Recommend posts that the similar user interacted with but the target user hasn't
        for post_id, interaction in similar_user_posts.items():
            if interaction > 0 and user_item_matrix.at[user_id, post_id] == 0:
                if post_id not in post_recommendations:
                    post_recommendations[post_id] = user_similarity_df.at[user_id, similar_user] * interaction
                else:
                    post_recommendations[post_id] += user_similarity_df.at[user_id, similar_user] * interaction

    recommended_post_ids = sorted(post_recommendations, key=post_recommendations.get, reverse=True)[:num_recommendations]

    return recommended_post_ids

user_item_matrix = create_user_item_matrix(unified_interactions, final_all_posts_df)
user_similarity_df = compute_user_similarity(user_item_matrix)

recommended_posts = recommend_posts(user_id=1, user_similarity_df=user_similarity_df, user_item_matrix=user_item_matrix, num_recommendations=5)
print("Recommended Posts for User 1:", recommended_posts)
```

# **HYBRID MODELS**

```python
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)
def recommend_hybrid(user_id, num_recommendations=5, alpha=0.5):
    # Get content-based recommendations
    content_based_posts = recommend_videos_for_user(user_id, num_recommendations)

    # Get collaborative filtering recommendations
    collaborative_posts_ids = recommend_posts(user_id, user_similarity_df, user_item_matrix, num_recommendations)
    collaborative_posts = final_all_posts_df[final_all_posts_df['id'].isin(collaborative_posts_ids)][['id', 'title']]
    collaborative_posts['collaborative_rank'] = collaborative_posts.index + 1
    collaborative_posts = collaborative_posts.reset_index(drop=True)

    # Add ranks for content-based posts
    content_based_posts['content_based_rank'] = content_based_posts.index + 1
    merged_recommendations = pd.merge(content_based_posts, collaborative_posts, on='id', how='left')

    merged_recommendations['content_based_rank'] = (
        (merged_recommendations['content_based_rank'] - merged_recommendations['content_based_rank'].min()) /
        (merged_recommendations['content_based_rank'].max() - merged_recommendations['content_based_rank'].min())
    )
    merged_recommendations['collaborative_rank'] = (
        (merged_recommendations['collaborative_rank'] - merged_recommendations['collaborative_rank'].min()) /
        (merged_recommendations['collaborative_rank'].max() - merged_recommendations['collaborative_rank'].min())
    )

    merged_recommendations['final_rank'] = (
        alpha * merged_recommendations['content_based_rank'].fillna(0) +
        (1 - alpha) * merged_recommendations['collaborative_rank'].fillna(0)
    )

    merged_recommendations = merged_recommendations.sort_values(by='final_rank')
    merged_recommendations['title'] = merged_recommendations['title_x'].fillna(merged_recommendations['title_y'])

    return merged_recommendations[['id', 'title', 'final_rank']].head(num_recommendations)
recommended_posts_hybrid = recommend_hybrid(user_id=1, num_recommendations=5)
print("Hybrid Recommendations for User 1:")
print(recommended_posts_hybrid)
```

# **COLD START**

```python
def preprocess_emotions(emotion_str):
    #clean it by removing unnecessary characters and converting to lowercase
    if isinstance(emotion_str, str):
        return emotion_str.lower().replace(',', ' ').replace(';', ' ').strip()
    return ''

# Cold Start Recommendation based on user mood
def recommend_based_on_mood(user_mood, final_all_posts_df, num_recommendations=5):
    # Preprocess the user's mood
    user_mood_processed = preprocess_emotions(user_mood)
    final_all_posts_df['processed_emotions'] = final_all_posts_df['emotions'].apply(preprocess_emotions)
    vectorizer = TfidfVectorizer(stop_words='english')
    all_emotions = final_all_posts_df['processed_emotions'].tolist() + [user_mood_processed]
    tfidf_matrix = vectorizer.fit_transform(all_emotions)

    user_mood_vector = tfidf_matrix[-1]
    post_emotions_matrix = tfidf_matrix[:-1]
    similarity_scores = cosine_similarity(user_mood_vector, post_emotions_matrix).flatten()me
    final_all_posts_df['similarity_score'] = similarity_scores
    recommended_posts = final_all_posts_df[['id', 'title', 'similarity_score']].sort_values(by='similarity_score', ascending=False)
    return recommended_posts[['id', 'title']].head(num_recommendations)

user_mood = "happy, excited, energetic"
recommended_posts_cold_start = recommend_based_on_mood(user_mood, final_all_posts_df, num_recommendations=5)

print("Cold Start Recommendations based on User Mood:")
print(recommended_posts_cold_start)
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error

def calculate_metrics(actual_ratings, predicted_ratings):
    # Convert actual_ratings to pandas Series if it's a NumPy array
    if isinstance(actual_ratings, np.ndarray):
        actual_ratings = pd.Series(actual_ratings)

    # Check if actual_ratings is empty or contains only NaNs
    if actual_ratings.empty or actual_ratings.isnull().all():
        return np.nan, np.nan # or some other appropriate value indicating no valid ratings

    mae = mean_absolute_error(actual_ratings, predicted_ratings)
    rmse = np.sqrt(mean_squared_error(actual_ratings, predicted_ratings))
    return mae, rmse


def evaluate_content_based(user_id, num_recommendations=5):
    recommended_posts = recommend_videos_for_user(user_id, num_recommendations)
    actual_ratings = unified_interactions[unified_interactions['post_id'].isin(recommended_posts['id'])]['rating_percent']
    actual_ratings = actual_ratings.set_axis(unified_interactions[unified_interactions['post_id'].isin(recommended_posts['id'])]['post_id'])
    common_post_ids = actual_ratings.index.intersection(recommended_posts['id'])
    actual_ratings = actual_ratings.loc[common_post_ids]

    # Get predicted ratings (similarity score) for the recommended posts, setting 'id' as index
    predicted_ratings = recommended_posts.merge(final_all_posts_df[['id', 'similarity_score']], on='id', how='left').set_index('id')['similarity_score']
    predicted_ratings = predicted_ratings.loc[common_post_ids]
    actual_ratings, predicted_ratings = actual_ratings.align(predicted_ratings, join='inner')


    # Calculate metrics only if both actual and predicted ratings are available
    if actual_ratings.empty or predicted_ratings.empty:
        return np.nan, np.nan
    else:
        return calculate_metrics(actual_ratings, predicted_ratings)

def evaluate_collaborative(user_id, num_recommendations=5):
    recommended_posts = recommend_posts(user_id, user_similarity_df, user_item_matrix, num_recommendations)
    actual_ratings = unified_interactions[unified_interactions['post_id'].isin(recommended_posts)]['rating_percent']
    predicted_ratings = [0.5] * len(actual_ratings)
    return calculate_metrics(actual_ratings, predicted_ratings)


def evaluate_hybrid(user_id, num_recommendations=5):
    recommended_posts = recommend_hybrid(user_id, num_recommendations)
    actual_ratings_df = unified_interactions[unified_interactions['post_id'].isin(recommended_posts['id'])]
    actual_ratings = actual_ratings_df['rating_percent']
    recommended_posts_with_ratings = recommended_posts[recommended_posts['id'].isin(actual_ratings_df['post_id'])]
    predicted_ratings = recommended_posts_with_ratings['final_rank']

    # Calculate metrics only if both actual and predicted ratings are available and have the same length
    if actual_ratings.empty or predicted_ratings.empty:
        return np.nan, np.nan
    else:
        min_len = min(len(actual_ratings), len(predicted_ratings))
        actual_ratings = actual_ratings.values[:min_len]
        predicted_ratings = predicted_ratings.values[:min_len]

        return calculate_metrics(actual_ratings, predicted_ratings)
user_id = 1
num_recommendations = 5

content_mae, content_rmse = evaluate_content_based(user_id, num_recommendations)
collaborative_mae, collaborative_rmse = evaluate_collaborative(user_id, num_recommendations)
hybrid_mae, hybrid_rmse = evaluate_hybrid(user_id, num_recommendations)

print(f"Content-Based MAE: {content_mae}, RMSE: {content_rmse}")
print(f"Collaborative Filtering MAE: {collaborative_mae}, RMSE: {collaborative_rmse}")
print(f"Hybrid MAE: {hybrid_mae}, RMSE: {hybrid_rmse}")
```

```python
from flask import Flask, request, jsonify

app = Flask(__name__)

@app.route('/feed', methods=['GET'])
def get_feed():
    username = request.args.get('username')
    category_id = request.args.get('category_id')
    mood = request.args.get('mood')

    user_id = final_all_users_df[final_all_users_df['username'] == username]['id'].values[0]


    if category_id and mood:
        # Endpoint 1: Category and Mood-Based Recommendations
        filtered_posts = final_all_posts_df[final_all_posts_df['category'] == category_id]
        recommendations_df = recommend_based_on_mood(mood, filtered_posts, num_recommendations=10)
        recommendations = recommendations_df.to_dict('records')

    elif category_id:
        # Endpoint 2: Category-Based Recommendations
        filtered_posts = final_all_posts_df[final_all_posts_df['category'] == category_id]
        recommendations_df = recommend_hybrid(user_id, num_recommendations=10)
        recommendations_df = recommendations_df[recommendations_df['id'].isin(filtered_posts['id'])]
        recommendations = recommendations_df.to_dict('records')
    elif username:
        # Endpoint 3: General Personalized Feed (with cold start handling)
        user_interactions = unified_interactions[unified_interactions['user_id'] == user_id]

        if user_interactions.empty:
            # Cold start: Recommend popular or diverse videos
            recommendations = final_all_posts_df.sort_values(by=['view_count', 'upvote_count'], ascending=False).head(10).to_dict('records')
        else:
            recommendations_df = recommend_hybrid(user_id, num_recommendations=10)
            recommendations = recommendations_df.to_dict('records')
    else:
        recommendations = []

    return jsonify(recommendations)

if __name__ == '__main__':
    app.run(debug=True, port=5000)
```
